# 🚀 Fivetran Teleport Sync – Complete Guide (Concept + Usage + Example)

---

## 📌 What is Fivetran Teleport Sync?

Fivetran Teleport Sync is a high-speed initial data loading mechanism.

- Fast initial sync
- Bulk data transfer
- Avoids row-by-row extraction

---

## 🧠 Simple Understanding

Normal Sync → Slow (row-by-row)
Teleport Sync → Fast (bulk load)

---

## ⚙️ Why Needed?

- Large data volume
- Reduce DB load
- Faster ingestion

---

## 🏗️ Architecture

Source DB → Snapshot → Cloud Storage → Warehouse

---

## 🔄 Steps

1. Snapshot creation
2. Upload to staging (S3/GCS/Azure)
3. Bulk load into warehouse
4. Switch to incremental

---

## 📊 Example

MySQL (10M rows)

Without Teleport → Slow
With Teleport → Fast

---

## 💻 Example Commands

mysqldump orders > orders.sql

aws s3 cp orders.sql s3://fivetran-stage/

COPY INTO orders
FROM @stage/orders.sql
FILE_FORMAT = (TYPE = 'CSV');
