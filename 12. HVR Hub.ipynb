# 🚀 HVR Hub (Fivetran HVR) – Complete Guide with Usage, Architecture & Examples

---

## 📌 What is HVR Hub?

HVR Hub is the central server component of HVR (High Volume Replication), now part of Fivetran.

- Manages replication jobs
- Stores metadata
- Orchestrates data movement

---

## 🧠 Simple Understanding

HVR Hub = Brain of replication system

---

## 🏗️ Architecture

Source DB → Capture → HVR Hub → Integrate → Target DB

---

## 🔑 Components

- Hub → Controller
- Capture → Reads changes
- Integrate → Applies changes
- Channel → Pipeline

---

## 📊 Use Cases

- Real-time replication
- Data warehouse loading
- Cloud migration
- Disaster recovery

---

## ⚙️ Commands

### Start Hub

hvrhubserver start

---

### Add Locations

hvraddloc my_channel src_mysql
hvraddloc my_channel tgt_snowflake

---

### Add Table

hvraddtable my_channel orders

---

### Activate

hvractivate -c my_channel
hvractivate -i my_channel

---

### Start Replication

hvrstart my_channel

---

### Compare Data

hvrcompare my_channel

---

### Restart

hvrrestart my_channel

---

### Refresh

hvrrefresh my_channel

---
