# 🚀 MAR (Monthly Active Rows) in Fivetran — COMPLETE DEEP DIVE

---

# 📌 1. What is MAR?

MAR = Monthly Active Rows

Number of unique rows that are inserted, updated, or deleted in a month

---

# 🧠 Core Idea

Pricing is based on changes in data, not total data

---

# 🔥 2. How MAR is Calculated

Example:

users table

id | name
1  | A
2  | B
3  | C

During month:

Insert id=4 → counted
Update id=1 → counted
Update id=1 again → not counted
No change id=2 → not counted

Final MAR = 2

---

# 🧠 Important Rules

- Unique rows only
- Same row counted once
- Only changes count
- Deletes may count

---

# 💰 3. Initial Load

First time full load

FREE (not counted in MAR)

Example:
10M rows → MAR = 0

---

# 🔁 4. Incremental Sync

Only changed rows

Example:
Day1: 100
Day2: 200

MAR = 300

---

# 🔄 5. Re-sync

Reload data again

FULL data counted again

Example:
1M rows → counted again

---

# ⚖️ 6. Comparison

Historical → Full → Free
Incremental → Changed → Paid
Re-sync → Full → Paid

---

# 🧠 7. Real Example

Table = 10M

Initial → Free
Monthly changes → 250K

Re-sync → 10M counted

---

# ⚠️ 8. Mistakes

- Frequent re-sync
- Not tracking MAR
- Dirty data

---

# 🚀 9. Best Practices

- Avoid re-sync
- Monitor MAR
- Use incremental
- Clean data

---

# 🎯 Summary

MAR = changed rows
Initial load = free
Re-sync = expensive
