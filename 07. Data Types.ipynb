# 🚀 Data Types in Fivetran
## Explicit vs Implicit + Data Type Hierarchy (Deep Explanation with Examples)

---

# 📌 1. EXPLICIT DATA TYPE (From Source Schema)

## Definition
Data type is defined by the source system schema

## Example

Step 1:
count → INT

Step 2:
count → DOUBLE

Step 3:
Supertype(INT, DOUBLE) = DOUBLE

Final:
count → DOUBLE

---

# 📌 2. IMPLICIT DATA TYPE (From Data)

## Definition
Data type is inferred from values

## Example

Step 1:
100 → DOUBLE

Step 2:
"ABC" → STRING

Step 3:
Supertype(DOUBLE, STRING) = STRING

Final:
count → STRING

# 🧠 Key Rules

- Always promote upward
- Never downgrade
- Choose widest type

---

# 🔄 Examples

1. INT + FLOAT → FLOAT
2. INT + DOUBLE → DOUBLE
3. INT + STRING → STRING

# ⚠️ Issues

- Mixed types → STRING
- Performance issues
- Wrong aggregations

# 🚀 Best Practices

- Avoid mixed types
- Use schema
