# 🚀 Sync Strategy in Fivetran
## Magic Sync vs Merge Sync

---

# 📌 Overview

Fivetran uses different **sync strategies** to load and update data in destination tables:

1. Magic Sync (Default)
2. Merge Sync (Advanced / Controlled)

---

# 🧠 Why Sync Strategy Matters

- How updates are handled
- How deletes are handled
- Performance
- Data consistency

---

# ✨ 1. MAGIC SYNC (Default)

---

## 🎯 What it is

Fully automated sync strategy used by Fivetran

- No configuration required
- Works out-of-the-box

---

## ⚙️ How it Works

1. Extract data from source
2. Load into staging table
3. Replace / update destination table
4. Maintain metadata columns

---

## 🧩 Key Features

- Automatic schema handling
- Handles inserts, updates, deletes
- Uses internal logic (black-box)

---

## 📌 Columns Added

- _fivetran_synced
- _fivetran_deleted
- _fivetran_id

---

## ✅ Advantages

- Simple
- No setup
- Works for most use cases

---

## ❌ Limitations

- Less control
- Can be inefficient for large tables
- Not customizable

---

## 🔥 Example

Source table updates → Fivetran auto sync → Destination updated

---

# 🔄 2. MERGE SYNC (Custom Strategy)

---

## 🎯 What it is

Uses SQL MERGE (UPSERT) logic in destination

---

## ⚙️ How it Works

MERGE INTO target t
USING staging s
ON t.id = s.id
WHEN MATCHED THEN UPDATE
WHEN NOT MATCHED THEN INSERT;

---

## 🧩 Key Features

- Controlled updates
- Uses primary key
- Efficient incremental loads

---

## 📌 Requirements

- Primary key must exist
- Destination supports MERGE

---

## ✅ Advantages

- Better performance
- Fine control over updates
- Efficient for large datasets

---

## ❌ Limitations

- Requires configuration
- Not available for all connectors
- More complex

---

## 🔥 Example

Update row id=1 → Only that row updated (not full reload)

---

# ⚖️ Magic vs Merge — Comparison

| Feature | Magic Sync | Merge Sync |
|--------|-----------|-----------|
| Setup | Automatic | Manual |
| Control | Low | High |
| Performance | Medium | High |
| Use case | General | Large/critical data |
| Logic | Internal | SQL MERGE |

---

# 🧠 When to Use What

## ✨ Use Magic Sync

- Small to medium data
- Quick setup
- No need for customization

---

## 🔄 Use Merge Sync

- Large tables
- Need efficiency
- Want control over updates

---

# 🧠 Internal Difference

## Magic Sync
- Fivetran decides everything
- May use truncate + reload or internal merge

## Merge Sync
- Explicit SQL merge logic
- Controlled row-level updates

---

# 🎯 Key Takeaway

Magic Sync is a fully automated default strategy, while Merge Sync provides controlled, efficient updates using SQL merge operations, especially useful for large datasets.

---

# 🔥 One-Line Interview Answer

Fivetran’s Magic Sync is an automated default sync strategy, whereas Merge Sync uses SQL-based upserts for efficient and controlled incremental data updates.
