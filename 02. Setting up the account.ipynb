# ☁️ STEP 1: Create Fivetran Account

## Steps

1. Go to Fivetran website
2. Click Start Free Trial / Sign Up
3. Sign up using Google / Email
4. Verify email

---

# ❄️ STEP 2: Create Snowflake Account

## Steps

1. Go to Snowflake website
2. Click Start Free Trial
3. Fill cloud + region

---

# STEP 3: Setup Snowflake

```sql
CREATE ROLE fivetran_role;

CREATE USER fivetran_user
PASSWORD = 'StrongPassword123'
DEFAULT_ROLE = fivetran_role;

GRANT ROLE fivetran_role TO USER fivetran_user;

CREATE WAREHOUSE fivetran_wh
WITH WAREHOUSE_SIZE = 'XSMALL'
AUTO_SUSPEND = 60;

GRANT USAGE ON WAREHOUSE fivetran_wh TO ROLE fivetran_role;

CREATE DATABASE fivetran_db;

GRANT ALL PRIVILEGES ON DATABASE fivetran_db TO ROLE fivetran_role;
```

---

# STEP 4: Connect in Fivetran

Fill:
- Host: <account>.snowflakecomputing.com
- DB: fivetran_db
- User: fivetran_user
- Password
- Warehouse: fivetran_wh
- Role: fivetran_role

Choose source → enter credentials → sync
