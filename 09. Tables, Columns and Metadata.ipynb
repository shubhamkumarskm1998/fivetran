# 🚀 How Fivetran Handles Tables, Columns, Deduplication & Metadata

---

# 📌 Overview

Fivetran automatically manages:

- Table names
- Column names
- Deduplication
- History tracking
- Metadata (_fivetran_* columns)

---

# 🧠 1. Table Name Handling

- Auto creates tables
- Normalizes names

Examples:
users → users
Orders → orders
sales-data → sales_data

Rules:
- Lowercase
- Replace special chars with _

---

# 🧩 2. Column Name Handling

Examples:
UserID → user_id
Order Date → order_date
Amount($) → amount

---

# 🔁 3. Deduplication

- Uses primary key
- If no PK → _fivetran_id

Flow:
1. Identify row
2. Keep latest version
3. Avoid duplicates

---

# 🗂️ 4. _fivetran Columns

_fivetran_synced → last sync time
_fivetran_deleted → soft delete flag
_fivetran_id → unique id

---

# 🧠 5. History Tracking

- Tracks updates
- Supports SCD-like behavior

---

# 🗄️ 6. Metadata

- Sync history
- Logs
- Schema changes

---

# 🔄 Flow

1. Extract
2. Normalize
3. Add metadata
4. Deduplicate
5. Load
6. Track

---

# ⚠️ Notes

- No hard delete
- Extra columns expected

---

# 🚀 Best Practices

- Use _fivetran_deleted
- Use _fivetran_synced
- Don’t remove metadata columns

---

# 🎯 Summary

Fivetran auto handles schema, deduplication, and history using metadata columns.
