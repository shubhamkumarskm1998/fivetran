# 🚀 Fivetran Web UI Overview
## Connectors, Transformations, Destinations & BI Options

---

# 📌 Overview

Fivetran Web UI is the control plane where you:
- Configure connectors
- Manage destinations
- Monitor syncs
- Enable transformations

---

# 🖥️ MAIN UI SECTIONS

Dashboard → Destinations → Connectors → Transformations → Logs/Monitoring

---

# 🧭 1. DASHBOARD

## What you see
- Active connectors
- Sync status
- Errors / alerts
- Data usage

## Key Metrics
- Last sync time
- Rows processed
- Connector health

---

# 🔌 2. CONNECTORS

## What is a Connector?
A pipeline that extracts data from source and loads into destination

## UI Features
- Add connector
- Pause / resume
- Force sync
- Schema selection

## Example
Google Sheets → Snowflake

## Settings
- Credentials
- Sync frequency
- Table selection

---

# ☁️ 3. DESTINATIONS

## What is a Destination?
Where data is loaded (warehouse)

## Supported
- Snowflake
- BigQuery
- Redshift
- Databricks

## Settings
- Database
- Schema
- Warehouse
- Role

## Managed Destination
Fivetran manages compute and infra

---

# 🔄 4. TRANSFORMATIONS

## What is Transformation?
Convert raw data into analytics-ready data

## Types
- SQL
- dbt

## Features
- Run transformations
- Monitor runs

---

# 📊 5. MONITORING

## Tracks
- Sync status
- Errors
- Latency

---

# 📈 6. BI TOOLS

Fivetran does NOT provide BI

Use:
- Tableau
- Power BI
- Looker

## Flow
Source → Fivetran → Snowflake → BI Tool

---

# 🎯 SUMMARY

Connector = Extract
Destination = Load
Transformation = Transform
BI = External tools
