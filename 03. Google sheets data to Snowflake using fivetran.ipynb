# 🚀 Load Data from Google Drive → Fivetran → Snowflake

---

# 📌 Overview

Goal:
Use ONLY Fivetran to create tables in Snowflake from Google Drive files

Google Drive / Sheets → Fivetran Connector → Snowflake Tables

---

# 🧠 Key Concept

Fivetran:
- Extracts data from Google Drive / Sheets
- Automatically creates tables in Snowflake
- Keeps data in sync

---

# 🚀 Steps

## STEP 1: Prepare File
- Use Google Sheets / CSV
- First row = column names
- Clean data

---

## STEP 2: Setup Snowflake
- Database ready
- Warehouse running
- Permissions set

---

## STEP 3: Add Connector
- Login to Fivetran
- Add Connector → Google Sheets
- Authenticate Google

---

## STEP 4: Configure
- Select file
- Select sheet
- Choose Snowflake destination

---

## STEP 5: Sync
- Start initial sync

---

## STEP 6: Verify

```sql
SELECT * FROM <schema>.<table>;
```

---

# 🧠 Internal Flow

1. Read data via API
2. Infer schema
3. Create table
4. Load data
5. Track changes

---

# 🔁 Incremental Sync

- Detects changes
- Updates only changed rows

---

# ⚠️ Issues

- Blank rows
- Mixed data types
- Missing headers

---

# 🎯 Summary

Fivetran automates ingestion from Google Drive to Snowflake without staging.
